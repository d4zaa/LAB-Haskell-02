```haskell
 divisores :: Int -> [Int]
 divisores a = [x | x <- [1..a], a `mod` x == 0]
 divisores 39
```

```haskell
divisoresPropios :: Int -> [Int]
divisoresPropios a = init [x | x <- [1..a], a `mod` x == 0]

esPerfecto :: Int -> Bool
esPerfecto a = sum (divisoresPropios a) == a

esPerfecto 6
esPerfecto 10
```
