```haskell
--con recursion
miLength :: [a] -> Int
miLength [] = 0
miLength (_:xs) = 1 + miLength xs

miLast :: [a] -> a
miLast [x] = x
miLast (_:xs) = miLast xs

duplicar :: [Int] -> [Int]
duplicar [] = []
duplicar (x:xs) = x*2 : duplicar xs

eliminarPares :: [Int] -> [Int]
eliminarPares [] = []
eliminarPares (x:xs)
    | even x = eliminarPares xs
    | otherwise = x : eliminarPares xs

--test
miLength [1,2,3,4]
miLast ["a", "b", "c", "d"]
duplicar [2,4,6,8]
eliminarPares [1, 3, 4, 5, 8]
```

```haskell
--con orden superior
miLenght :: [a] -> Int
miLenght = length

miLast :: [a] -> a
miLast = last

duplicar :: [Int] -> [Int]
duplicar = map (*2) 

eliminarPares :: [Int] -> [Int]
eliminarPares = filter odd 

--test
miLength [1,2,3,4]
miLast ["a", "b", "c", "d"]
duplicar [2,4,6,8]
eliminarPares [1, 3, 4, 5, 8]
```
